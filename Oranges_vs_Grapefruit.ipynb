Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Continuando com o projeto das laranjas/toranjas agora você deve calcular o coeficiente ângular e o linear para a reta da laranja e para a reta da toranja. Use a fórmula de mínimos quadrados para encontrar cada um.

```python
x = diametro_laranjas
y = peso_laranjas
n = peso_laranjas.size
```

```python
a = (n*sum(x*y) - sum(x)*sum(y))/(n*sum(x**2) - (sum(x)**2))
b = np.mean(y) - a*np.mean(x)
Y = a*x+b
print((a,b))
```

```python
plt.plot(x, y)
plt.plot(x, Y)
```

```python
x = diametro_toranjas
y = peso_toranjas
n = peso_toranjas.size
```

```python
a = (n*sum(x*y) - sum(x)*sum(y))/(n*sum(x**2) - (sum(x)**2))
b = np.mean(y) - a*np.mean(x)
Y = a*x+b
print((a,b))
```

```python
plt.plot(x, y)
plt.plot(x, Y)
```

4.
Chegou a hora de você testar os conhecimentos desenvolvidos durante a aula. Continuando com o projeto das laranjas/toranjas, agora você deve calcular o coeficiente angular utilizando a geração de números aleatórios. Assuma que já conhece b e que este é igual a 17.

```python
x1 = diametro_laranjas
y1 = peso_laranjas
norma1 = np.array([])
np.random.seed(69)
coef_angulares1 = np.random.uniform(low = 13, high = 14.5, size = 5000)
for i in range(5000):
  norma1 = np.append(norma1, np.linalg.norm((coef_angulares1[i]*x1 + 17)- y1))
  if norma1[i] == min(norma1):
    a1 = coef_angulares1[i]
Y1 = a1*x1 + 17
```

```python
plt.plot(x1, y1)
plt.plot(x1, Y1)
```

```python
dados1 = np.column_stack([coef_angulares1, norma1])
np.savetxt('dados_reg_laranjas.csv', dados, delimiter = ',')
```
